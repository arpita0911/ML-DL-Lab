#	Multiple Linear Regression

Housing price prediction using the Boston Housing Dataset by capturing linear relationships between multiple input features and the target value.

```python
import numpy as np
import pandas as pd

from tensorflow.keras.datasets import boston_housing

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
def evaluate_model(name, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n{name}")
    print(f"MSE: {mse:.4f}")
    print(f"R² Score: {r2:.4f}")
```

```python
def multiple_linear_regression(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    evaluate_model("Multiple Linear Regression", y_test, y_pred)
```

```python
multiple_linear_regression(X_train, X_test, y_train, y_test)
```

The model explains about 72.14% of the variation in housing prices (R² = 0.7214) and has a much lower error of 23.20 (MSE), showing a significant improvement in prediction accuracy with multiple features.
